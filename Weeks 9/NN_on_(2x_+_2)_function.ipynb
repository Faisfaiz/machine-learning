## **NN on (2x + 2) function**



```python
# Install TensorFlow
# pip install tensorflow

```

```python
# Import library yang dibutuhkan
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fungsi untuk menghasilkan data latih
def generate_data():
    X = np.random.uniform(-10, 10, 1000)  # Data latih pada rentang -10 hingga 10
    y = 2 * X + 2 + np.random.normal(0, 2, 1000)  # Fungsi 2x + 2 dengan noise

    return X, y

# Membuat data latih
X_train, y_train = generate_data()
```

```python
# Membangun model jaringan saraf
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=(1,))
])
```

```python
# Mengompilasi model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Latih model dan simpan riwayat pelatihan
history = model.fit(X_train, y_train, epochs=50)
```

```python
# Plot sejarah pelatihan
plt.plot(history.history['loss'])
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.show()
```

```python
# Melakukan prediksi menggunakan model yang telah dilatih
X_test = np.linspace(-5, 5, 100).reshape(-1, 1)
y_pred = model.predict(X_test)
```

```python
# Plot fungsi asli dan nilai yang diprediksi
plt.scatter(X_train, y_train, label='Training Data')
plt.plot(X_test, y_pred, color='red', label='Predicted Function: 2x + 2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
