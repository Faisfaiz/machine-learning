```python
import pandas as pd
```

```python
dataset = "melb_data.csv"
df = pd.read_csv(dataset) # Membaca file csv
df.head() # Untuk menampilkan data sebanyak ()
```

```python
print(df.columns)
```

```python
df = df.dropna() # Untuk mengedrop colom yang bernilai kosong (NaN)
```

```python
y = df.Price
print(y)
```

```python
Features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
x = df[Features]
x.describe() # Menggunakan fungsi describe
```

```python
x.head() # Untuk menampilkan data
```

```python
from sklearn.tree import DecisionTreeRegressor

Model = DecisionTreeRegressor(random_state=1)
Model.fit(x, y)
```

```python
print("Making predictions for the following 5 houses:")
print(x.head())
print("The predictions are")
print(Model.predict(x.head()))
```
